#### Task 3: User Menu(while loop + break/continue)

1. Implement a simple while loop driven menu that keeps asking the user to choose an action until they press q to quit
Menu options:
1 - Add order amount to running list(use input())
2 - Show all orders and totals after applying discounts
3 - Quit(Press q)

2. Use continue to re-show the menu after invalid input and break to exit on q

```python
orders_list = []

while True:
    print("------MENU-------")
    print("1.Add order amount to running list")
    print("2 - Show all orders and totals after applying discounts")
    print("Press q to quit")

    choice = input("Enter any options from menu")

    if choice =="1":
        amount = float(input("Enter any order amount"))
        orders_list.append(amount)
        print("The order amount was successfully added to order list")

    elif choice == "2":
        if not orders_list:
            print("No available orders")

        else:
            total_revenue = 0
            print("Orders","  ","Discount%","  ","Final Amount")
            for order_amount in orders_list:
                if order_amount >= 2000:
                    discount_val = 15

                elif 1500 <= order_amount < 2000:
                    discount_val = 10

                elif 1000 <= order_amount < 1500:
                    discount_val = 7

                else:
                    discount_val = 0

                final_amount = ((100 - discount_val)/100)*order_amount
                print(order_amount,"->",discount_val,"->",final_amount)

                total_revenue += final_amount 

            print("Total revenue after the discount is:",total_revenue)

    elif choice == "q":
        print("End of the program....")
        break 

    else:
        print("Invalid choice")
        continue
```

#### Task 4: Loop control with conditions:

1. Given a list of daily sales: daily = [200,150,0,400,50,-1,300], iterate through it with a for loop and:
   - if a day's value is -1, treat it as corrupted data and break the loop(stop processing)
   - if a day's value is 0, treat it as a day with no sales and continue(skip adding revenue)
   - for valid positive sales, add to total_sales and print the running total  

2. Print the final total after the loop completes(or stops due to corrupted data)

```python
daily = [200,150,0,400,50,-1,300]
total_sales = 0

for sales in daily:
    if sales == -1:
        print("Corrupt data...")
        break 
    elif sales == 0:
        print("No sales...")
        continue
    else:
        total_sales += sales
        print(f"Running total for sales is: {total_sales}")

print(f"The total sales is: {total_sales}")
```
